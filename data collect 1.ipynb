```python
from PIL import Image
import pytesseract
import glob
import os

# Point to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Configuration: Sinhala language
custom_config = '--psm 6 -l sin'

# Get all generated image files that match pattern
image_files = sorted(glob.glob('page_*.png'))

# Loop through and extract text
all_text = []

for image_path in image_files:
    print(f"Processing {image_path}...")
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img, config=custom_config)
    all_text.append(text)

# Print combined result
full_text = "\n".join(all_text)
print("\n\n=== Extracted Sinhala Text ===\n")
print(full_text)

# Optional: Save to a file
with open("extracted_sinhala_text.txt", "w", encoding="utf-8") as f:
    f.write(full_text)
```

```python
import re

def extract_numbered_questions(text):
    # Normalize line breaks
    text = text.replace('\r\n', '\n').replace('\r', '\n')

    # Regex to find numbered blocks (e.g., "1. What is...?")
    pattern = re.compile(r'(\d{1,3}\..*?)(?=\n\d{1,3}\.|\Z)', re.DOTALL)
    
    matches = pattern.findall(text)
    
    cleaned_questions = []
    for match in matches:
        # Remove extra spaces and line breaks
        cleaned = re.sub(r'\s+', ' ', match).strip()
        
        # Ensure it ends with '?' and is a proper question
        if cleaned.endswith('?'):
            # Optional: Check if it starts with a question word (e.g., "What", "How", "Why")
            # (Adjust for Sinhala question words if needed)
            question_words = ["තෝරන්න", "කුමක්‌ ද", "වූයේ", "කාලයේදීද", "ලියන්න", "කරන්න", "which", "ඇයි", "කොහෙද", "කවුද", "කෙනෙක්", "මොකක්ද"]  # Add Sinhala equivalents
            
            # Case-insensitive check for question words
            if any(cleaned.lower().split()[1].startswith(word) for word in question_words):
                cleaned_questions.append(cleaned)
            else:
                # If no question word, still keep if it ends with '?'
                cleaned_questions.append(cleaned)
    
    return cleaned_questions

# Extract
questions = extract_numbered_questions(full_text)

# Show and save
print("\n=== Valid Sinhala Questions ===\n")
for i, q in enumerate(questions, 1):
    print(f"{i}. {q}\n")

with open("cleaned_sinhala_questions.txt", "w", encoding="utf-8") as f:
    for q in questions:
        f.write(q + "\n")
```

```python
# Regular expression pattern to match questions (two digits followed by a dot)
question_pattern = re.compile(r'^\d{2}\..+')

all_questions = []

for image_path in image_files:
    print(f"Processing {image_path}...")
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img, config=custom_config)
    
    # Split text into lines and find questions
    for line in text.split('\n'):
        if question_pattern.match(line.strip()):
            all_questions.append(line.strip())

# Print all found questions
print("\n\n=== Extracted Questions ===\n")
for i, question in enumerate(all_questions, 1):
    print(f"{i}. {question}")

# Save questions to a file
with open("extracted_questions.txt", "w", encoding="utf-8") as f:
    for i, question in enumerate(all_questions, 1):
        f.write(f"{i}. {question}\n")
```
